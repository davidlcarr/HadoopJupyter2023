# Projeto Fase 2 - Solu√ß√£o de BI para Gest√£o de Notas Fiscais (SILOMS)

Este reposit√≥rio cont√©m o c√≥digo-fonte, documentos e dashboards gerados como parte da Fase 2 do projeto de Business Intelligence, com foco na an√°lise e monitoramento de vencimentos de notas fiscais extra√≠das do sistema SILOMS.

## üéØ Objetivo
Criar uma solu√ß√£o automatizada de BI que permita acompanhar o status de vencimento das notas fiscais, gerar KPIs relevantes e fornecer relat√≥rios visuais consolidados para apoio √† tomada de decis√£o.

## üîÅ Pipeline da Solu√ß√£o

1. **Importa√ß√£o da planilha `.xlsx`** extra√≠da do sistema SILOMS.
2. **Tratamento dos dados** com Python (limpeza, formata√ß√£o de datas, extra√ß√£o de nomes).
3. **C√°lculo de vencimentos** (30 dias ap√≥s a emiss√£o da nota).
4. **Classifica√ß√£o das notas** como `Em dia` ou `Atrasadas`.
5. **Gera√ß√£o de KPIs e gr√°ficos** (status, distribui√ß√£o, m√©dia de dias).
6. **Exporta√ß√£o para Excel** com destaque em vermelho para notas vencidas.
7. **Gera√ß√£o de PDF** com dashboard consolidado (tabela + gr√°ficos).

````python
# üìå 1. Importa√ß√£o de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from openpyxl import load_workbook
from openpyxl.styles import PatternFill
from fpdf import FPDF
````

````python
# üìå 2. Carregamento e prepara√ß√£o dos dados
# Fa√ßa upload do arquivo .xlsx manualmente no Colab
from google.colab import files
uploaded = files.upload()
````

````python
# üìå Substitua pelo nome correto, se necess√°rio
df = pd.read_excel("RelGestorFin.xlsx", skiprows=1)

# üìå Renomear colunas
df.columns = [
    "nota_fiscal", "status", "empresa", "data_emissao", "data_entrada", "ug_cred", "ug_pagadora",
    "ug_atual", "situacao_atual", "data_situacao", "dias_na_situacao", "np", "ne", "nd",
    "ptres", "fonte_recurso", "valor_fat_emp", "valor_total_nf"
]

# üìå Remover colunas desnecess√°rias
colunas_excluir = [
    "status", "ug_cred", "ug_pagadora", "ug_atual",
    "data_situacao", "np", "ptres", "fonte_recurso", "valor_fat_emp", "ne", "nd"
]
df.drop(columns=colunas_excluir, inplace=True)

# üìå Ajustar a coluna "empresa"
# Remove tudo que vem antes e incluindo '|', e mant√©m s√≥ a primeira palavra
df["empresa"] = df["empresa"].astype(str).str.split("|").str[-1].str.strip()
df["empresa"] = df["empresa"].str.split().str[0]

# üìå Convers√£o de datas
#df["data_emissao"] = pd.to_datetime(df["data_emissao"], errors="coerce")
#df["data_entrada"] = pd.to_datetime(df["data_entrada"], errors="coerce")

# üìå Formatando datas no padr√£o brasileiro (dd/mm/aaaa)
df["data_emissao"] = df["data_emissao"].dt.strftime("%d/%m/%Y")
df["data_entrada"] = df["data_entrada"].dt.strftime("%d/%m/%Y")
````

````python
# üìå Converter datas
df["data_emissao"] = pd.to_datetime(df["data_emissao"], errors="coerce")

# üìå Calcular vencimento (30 dias ap√≥s emiss√£o)
df["data_vencimento"] = df["data_emissao"] + pd.Timedelta(days=30)

# üìå Calcular dias at√© o vencimento
hoje = pd.to_datetime(datetime.today().date())
df["dias_ate_vencimento"] = (df["data_vencimento"] - hoje).dt.days

# üìå Adicionar status de vencimento
df["status_vencimento"] = df["dias_ate_vencimento"].apply(lambda x: "Atrasada" if x < 0 else "Em dia")

# üìå Formatar texto do vencimento
def formatar_vencimento(dias):
    if pd.isna(dias):
        return "Data inv√°lida"
    elif dias > 0:
        return f"{dias} dias at√© o venc."
    elif dias == 0:
        return "Vence hoje"
    else:
        return f"{abs(dias)} dias ap√≥s o venc."

df["dias_ate_vencimento"] = df["dias_ate_vencimento"].apply(formatar_vencimento)

# üìå Formatar a data no padr√£o brasileiro
df["data_emissao"] = df["data_emissao"].dt.strftime("%d/%m/%Y")

# üìå Selecionar colunas na ordem desejada
final = df[[
    "nota_fiscal", "empresa", "data_emissao", "status_vencimento",
    "dias_ate_vencimento"
]]

final
````

````python
# Agrupar e contar quantidade de notas por empresa
contagem = final.groupby('empresa').size().sort_values(ascending=True)

# Configurar estilo clean
sns.set_context("notebook")
sns.set_style("white")  # sem linhas de grade fortes
cores = sns.color_palette("pastel", len(contagem))

# Criar figura
plt.figure(figsize=(12, 6))
bars = plt.barh(contagem.index, contagem.values, color=cores)

# R√≥tulos sutis no centro da barra
for bar in bars:
    largura = bar.get_width()
    plt.text(largura + 0.2, bar.get_y() + bar.get_height() / 2,
             str(int(largura)), va='center', ha='left', fontsize=9, color='#333333')

# Ajustes visuais
plt.title("Quantidade de Notas por Empresa", fontsize=13, weight='bold', loc='left')
plt.xlabel("Qtd. de Notas", fontsize=11)
plt.ylabel("")  # Remover r√≥tulo vertical

# Remover bordas e fundo
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')

# Grid leve no eixo x
ax.xaxis.grid(True, color='#EEEEEE')
ax.yaxis.grid(False)

plt.tight_layout()
plt.show()
````

````python
# Estilo visual
plt.style.use('ggplot')  # Estilo mais limpo e moderno

# Cores personalizadas
cores = {
    "Em dia": "#66bb6a",     # Verde
    "Atrasada": "#ef5350"    # Vermelho
}

# Contagem de status
total = len(final)
atrasadas = len(final[final["status_vencimento"] == "Atrasada"])
em_dia = total - atrasadas

# Gr√°fico de pizza: propor√ß√£o de notas
labels = ['Em dia', 'Atrasada']
sizes = [em_dia, atrasadas]
colors = [cores[label] for label in labels]

fig, ax = plt.subplots(figsize=(7, 7))
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops=dict(width=0.4, edgecolor='w'),
    textprops=dict(color="black", fontsize=12)
)

# Anota√ß√µes no meio
ax.set(aspect="equal", title='Distribui√ß√£o de Notas Fiscais por Status de Vencimento')
plt.setp(autotexts, size=14, weight="bold")
plt.tight_layout()
plt.show()

# Fun√ß√£o para extrair n√∫mero de dias
def extrair_dias(valor):
    try:
        return int(valor.split(" ")[0])
    except:
        return None

# Convers√£o dos textos da coluna
final["dias_ate_vencimento_num"] = final["dias_ate_vencimento"].apply(extrair_dias)

# Separar listas para gr√°fico de distribui√ß√£o
dias_atrasadas = final[final["status_vencimento"] == "Atrasada"]["dias_ate_vencimento_num"].dropna()
dias_em_dia = final[final["status_vencimento"] == "Em dia"]["dias_ate_vencimento_num"].dropna()

# Gr√°fico de distribui√ß√£o (histograma com densidade)
plt.figure(figsize=(12, 6))
sns.histplot(dias_em_dia, bins=15, color=cores["Em dia"], label="Em dia", kde=True, stat="count", alpha=0.6)
sns.histplot(dias_atrasadas, bins=15, color=cores["Atrasada"], label="Atrasada", kde=True, stat="count", alpha=0.6)
plt.axvline(0, color='gray', linestyle='--')
plt.xlabel('Dias at√©/ap√≥s o vencimento')
plt.ylabel('Quantidade de Notas')
plt.title('Distribui√ß√£o dos Dias at√©/ap√≥s Vencimento das Notas Fiscais')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Agrupa por empresa e status de vencimento
resumo_status = final.groupby(["empresa", "status_vencimento"])\
                     .size().reset_index(name="quantidade")

# Seleciona as top 10 empresas com mais notas totais
top_empresas = final["empresa"].value_counts().head(10).index
resumo_status = resumo_status[resumo_status["empresa"].isin(top_empresas)]

# Gr√°fico
plt.figure(figsize=(12, 6))
sns.barplot(data=resumo_status,
            x="empresa", y="quantidade", hue="status_vencimento",
            palette={"Em dia": "#66bb6a", "Atrasada": "#ef5350"})
plt.xlabel("Empresa")
plt.ylabel("Quantidade de Notas")
plt.title("Distribui√ß√£o de Notas por Empresa e Status de Vencimento")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Status")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

````

````python
import matplotlib.pyplot as plt

# Filtra apenas notas atrasadas com dias v√°lidos
atrasadas = final[(final["status_vencimento"] == "Atrasada") &
                  (final["dias_ate_vencimento_num"].notna())]

# Agrupa por empresa e calcula m√©dia de atraso
media_atraso = atrasadas.groupby("empresa")["dias_ate_vencimento_num"].mean().reset_index()
media_atraso = media_atraso.sort_values(by="dias_ate_vencimento_num", ascending=False).head(10)

# Gr√°fico
plt.figure(figsize=(12, 6))
bars = plt.barh(media_atraso["empresa"], media_atraso["dias_ate_vencimento_num"], color="#ef5350")

# Adiciona o valor da m√©dia no centro da barra
for bar in bars:
    largura = bar.get_width()
    plt.text(largura / 2,                          # metade da largura da barra
             bar.get_y() + bar.get_height() / 2,   # centro vertical
             f"{largura:.1f} dias",                # exibe com 1 casa decimal
             ha='center', va='center', fontsize=10, color='white', weight='bold')

# Estilo
plt.xlabel("M√©dia de Dias de Atraso")
plt.title("M√©dia de Atraso por Empresa (Notas Atrasadas)")
plt.gca().invert_yaxis()
plt.grid(axis="x", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
````

````python
# üìå Caminho para salvar o Excel
excel_path = "Resumo_Vencimentos.xlsx"

# üìå Exportar para Excel
final.to_excel(excel_path, index=False)

# üìå Aplicar formata√ß√£o no Excel com openpyxl
wb = load_workbook(excel_path)
ws = wb.active
vermelho = PatternFill(start_color="FFC7CE", end_color="FFC7CE", fill_type="solid")

# Percorrer linhas e pintar as que estiverem com status "Atrasada"
for row in range(2, ws.max_row + 1):
    if ws[f"D{row}"].value == "Atrasada":
        for col in range(1, ws.max_column + 1):
            ws.cell(row=row, column=col).fill = vermelho

wb.save(excel_path)
wb.close()

# üìå Gerar PDF com os 20 primeiros registros da tabela "final"
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=10)
pdf.cell(200, 10, txt="Resumo de Notas Fiscais e Vencimentos", ln=1, align="C")
pdf.ln(5)

# Cabe√ßalhos
headers = ["Nota Fiscal", "Empresa", "Data Emiss√£o", "Status", "Descri√ß√£o"]
col_widths = [30, 30, 30, 30, 70]

for i, header in enumerate(headers):
    pdf.cell(col_widths[i], 10, header, border=1)
pdf.ln()

# Adicionar os dados (m√°ximo 20 linhas)
for _, row in final.head(20).iterrows():
    pdf.cell(col_widths[0], 10, str(row["nota_fiscal"]), border=1)
    pdf.cell(col_widths[1], 10, str(row["empresa"]), border=1)
    pdf.cell(col_widths[2], 10, str(row["data_emissao"]), border=1)
    pdf.cell(col_widths[3], 10, str(row["status_vencimento"]), border=1)
    pdf.cell(col_widths[4], 10, str(row["dias_ate_vencimento"])[:40], border=1)
    pdf.ln()

# üìå Salvar o PDF
pdf.output("Resumo_Vencimentos.pdf")
````
